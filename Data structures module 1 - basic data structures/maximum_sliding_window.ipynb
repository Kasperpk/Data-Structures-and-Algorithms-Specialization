```python
def max_sliding_window(n, integers, m):

    max_values = []

    i = 0
    j = i+m-1
    while j < len(integers):
        max = 0
        for num in integers[i:j+1]:
            if num > max:
                max = num
        max_values.append(max)
        i = i + 1
        j = i+m-1

    return max_values
```

```python
max_sliding_window(8, [2,7,3,1,5,2,6,2],4)
```

```python
def max_sliding_window(n, integers, m):
    max_values = []
    
    for i in range(n - m + 1):
        window_max = max(integers[i:i+m])
        max_values.append(window_max)
    
    return max_values
```

```python
max_sliding_window(8, [2,7,3,1,5,2,6,2],4)
```

```python
from collections import deque

def max_sliding_window_optimized(n, integers, m):
    dq = deque()  # stores indices
    result = []
    
    for i in range(n):
        # Remove indices outside current window
        while dq and dq[0] <= i - m:
            dq.popleft()
        
        # Remove indices of elements smaller than current
        while dq and integers[dq[-1]] <= integers[i]:
            dq.pop()
        
        dq.append(i)
        
        # Add to result when window is complete
        if i >= m - 1:
            result.append(integers[dq[0]])
    
    return result
```

```python
max_sliding_window_optimized(8, [2,7,3,1,5,2,6,2],4)
```
