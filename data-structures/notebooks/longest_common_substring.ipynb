```python
def longest_common_substring(s,t):
    prefix = 0
    suffix = 0
    base = 31
    mod = 10**9 + 7
    power = 1
    last_index = -1

    for i in range(min(len(s), len(t))):
        prefix = (prefix * base + ord(s[i])) % mod
        print(prefix)
        suffix = (suffix + ord(t[len(t) - 1 - i]) * power) % mod
        print(suffix)
        if prefix == suffix:
            last_index = i
        power = (power * base) % mod
    if last_index == -1:
        return 0, 0, 0
    return last_index + 1, prefix, suffix
```

```python
if __name__ == '__main__':
    s = str(input())
    t = str(input())
    length, _, _ = longest_common_substring(s, t)
    print(length)
```
