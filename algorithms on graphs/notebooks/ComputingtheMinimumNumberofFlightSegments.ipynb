```python
def breadth_first_search(vertices, v, u):

    # if the vertex v is not in the list of vertices, then solution can't be found
    if v not in vertices.keys():
        return -1
    
    adj_list = {i: [] for i in range(1, m+1)}
    for vertex, neighbors in vertices.items():
        if vertex not in adj_list:
            adj_list[vertex] = []
        for neighbor in neighbors:
            if neighbor not in adj_list[vertex]:
                adj_list[vertex].append(neighbor)
            if vertex not in adj_list[neighbor]:
                adj_list[neighbor].append(vertex)
    distances = {}
    for vertex in vertices.keys():
        distances[vertex] = float('inf')

    distances[v] = 0

    def find_distance(vertex):
        for neighbor in adj_list[vertex]:
            if distances[neighbor] == float('inf'):
                distances[neighbor] = distances[vertex] + 1
                find_distance(neighbor)

    find_distance(v)
        
    # if u couldn't be reached from vertex or any neighbor then return -1
    if distances[u] == float('inf'):
        return -1
    
    else:
        return distances[u]
    

    
    
    
```

```python
n,m = map(int, input().split())

vertices = {i: [] for i in range(1,n+1)}

for i in range(0,m):
    a,b = map(int, input().split())
    vertices[a].append(b)

v,u = map(int, input().split())

print(breadth_first_search(vertices,v,u))
```
