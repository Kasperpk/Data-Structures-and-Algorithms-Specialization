```python
def is_bipartite(graph):
    color = {}
    
    def dfs(node, c):
        if node in color:
            return color[node] == c
        color[node] = c
        return all(dfs(nei, c ^ 1) for nei in graph[node])
    
    for vertex in graph:
        if vertex not in color:
            if not dfs(vertex, 0):
                return 0
    return 1

# Example usage:
if __name__ == "__main__":
    n,m = map(int, input().split())
    graph = {i: [] for i in range(1,n+1)}
    for _ in range(m):
        u,v = map(int, input().split())
        graph[u].append(v)
        graph[v].append(u)
    print(is_bipartite(graph))  # Output: 1
```
